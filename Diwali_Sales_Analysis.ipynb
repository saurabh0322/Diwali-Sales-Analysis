# Exploratory Data Analysis




### Gender

```python
df.columns
```

```python
ax = sns.countplot(
    x='Gender',
    data=df,
    hue='Gender',
    palette=['#FF69B4', '#1E90FF'],
    legend=False
)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax = sns.barplot(
    x='Gender',
    y='Amount',
    data=sales_gen,
    hue='Gender',
    palette=['#4C72B0', '#DD8452'],
    legend=False
)
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

### Age

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

***From above graphs we ca see that most of the buyers are of age group between 26-35 years female***

### State

```python
df.columns
```

```python
# Total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,6)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# total amount/sales from top 10  states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,6)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

***From above the graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and karnataka respectively***

### Marital Status

```python
ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status',
                   palette=['#4C72B0','#DD8452'], legend=False)
for c in ax.containers: ax.bar_label(c)



sns.set(rc={'figure.figsize':(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(4,4)})
sns.barplot(data=sales_state, x='Marital_Status',y='Amount',hue='Gender')
```

***From above graphs we can see that most of the buyers are married (women) and they have high purchasing power***

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(
    data=df,
    x='Occupation',
    hue='Occupation',
    palette='tab20',
    legend=False
)

for c in ax.containers:
    ax.bar_label(c)
```

```python
sns.set(rc={'figure.figsize':(20,7)})

ax = sns.barplot(
    data=sales_state,
    x='Occupation',
    y='Amount',
    hue='Occupation',
    palette='tab20',
    legend=False
)
```

***From above graphs we can see that most of the buyers are working in IT,Healthcare and Aviation sector***

### Product Category

```python
sns.set(rc={'figure.figsize':(30,10)})

ax = sns.countplot(
    data=df,
    x='Product_Category',
    hue='Product_Category',
    palette='tab10',
    legend=False
)

for c in ax.containers:
    ax.bar_label(c)
```

```python
sales_state = (
    df.groupby(['Product_Category'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
      .head(10)
)

sns.set(rc={'figure.figsize':(20,5)})

ax = sns.barplot(
    data=sales_state,
    x='Product_Category',
    y='Amount',
    hue='Product_Category',
    palette='tab10',
    legend=False
)
```

***From above graphs we can see the most of the sold product are from food,Clothing and Electronics category*** 

```python
import matplotlib.pyplot as plt
import seaborn as sns

sales_product = (
    df.groupby('Product_ID', as_index=False)['Orders']
      .sum()
      .sort_values(by='Orders', ascending=False)
      .head(10)
)

sns.set(rc={'figure.figsize':(12,5)})  # figsize using rc

ax = sns.barplot(
    data=sales_product,
    x='Product_ID',
    y='Orders',
    hue='Product_ID',
    palette='tab10',
    legend=False
)
```

```python
import matplotlib.pyplot as plt

# Top 10 products by Orders
top10 = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Plot
fig1, ax1 = plt.subplots(figsize=(12,8))
top10.plot(kind='bar', color=plt.cm.tab10.colors)  # different color per bar

ax1.set_title('Top 10 Most Sold Products', fontsize=16)
ax1.set_xlabel('Product ID', fontsize=12)
ax1.set_ylabel('Total Orders', fontsize=12)
plt.show()
```
